## Extracting mobile data from 100 pages in Flipkart

```python
names = []
price = []
links = []

flipkart_link = "https://www.flipkart.com"
main_link = "http://www.flipkart.com/mobiles/pr?p%5B%5D=sort%3Dprice_asc&sid=tyy%2C4io&layout=grid&ajax=true&start="

for i in range(0, 100):
    link = main_link + str(i)
    page = requests.get(link)
    soup = BeautifulSoup(page.text, "html.parser")
    for main_block in soup.find_all(name = "div", attrs= {"class":"_1HmYoV _35HD7C"}):
        for product_block in main_block.find_all(name="div", attrs={"class":"bhgxx2"}):
            for title_block in product_block.find_all(name ="div", attrs ={"class":"_3wU53n"}):
                print(title_block.text)
                if title_block.text:
                    names.append(title_block.text)
                else:
                    name.append("Not found")
            for link in product_block.find_all(name = "a", href=True, attrs={"class":"_31qSD5"}):
                print(link.get('href'))
                if link.get('href'):
                    links.append(flipkart_link + link.get('href'))
                else:
                    links.append("Not found")
            for price_block in product_block.find_all(name = "div", attrs = {"class":"_1vC4OE _2rQ-NK"}):
                print(price_block.text)
                if price_block.text:
                    print(price_block.text)
                    price.append(price_block.text)
                else:
                    price.append("Not found")
```

```python
flipkart_data['Mobile Name'] = names
flipkart_data['Price'] = price
flipkart_data['Link'] = links
```
